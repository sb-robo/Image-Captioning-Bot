```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
def readTextFile(path):
    with open(path,encoding='utf-8') as file:
        caption = file.read()
        
    return caption
```

```python
# read data for ecoded images

train_file = readTextFile('../Dataset/Flickr8k/Flickr_TextData/Flickr_8k.trainImages.txt')
test_file = readTextFile('../Dataset/Flickr8k/Flickr_TextData/Flickr_8k.testImages.txt')

train = train_file.split('\n')[:-1]
test = test_file.split('\n')[:-1]

print(len(train),len(test))
```

```python
with open('./Caption_description.pkl','rb') as f:
    Caption_dict = pickle.load(f)
```

- `Encode Images`

```python
import cv2
from keras.models import Model
from keras.preprocessing import image
from keras.applications.resnet50 import ResNet50,preprocess_input,decode_predictions
```

```python
resnet = ResNet50(weights='imagenet',input_shape=(224,224,3))
resnet.summary()
```

```python
model = Model(resnet.input,resnet.layers[-2].output)
```

```python
image_path = '../Dataset/Flickr8k/Images'

def process_img(img_name):
    img = image.load_img(os.path.join(image_path,img_name),target_size=(224,224))
    img = image.img_to_array(img)
    img = np.expand_dims(img,axis=0)
    
    #normalization of image
    img = preprocess_input(img)
    return img
```

```python
img = process_img(list(Caption_dict.keys())[0])
plt.axis('off')
plt.imshow(img[0])
plt.show()
```

```python
def encode_image(img):
    img = process_img(img)
    feature_vector = model.predict(img)
    feature_vector = feature_vector.reshape((-1,))
    #print(feature_vector.shape)
    return feature_vector
```

```python
encoding_train={}
#image_id --- > feature vector

for ix,img_name in enumerate(train):
    encoding_train[img_name] = encode_image(img_name)
    
    if (ix+1)%100 ==0:
        print(f'Complete: {ix+1} of {len(train)}')
```

```python
# Store in directoty
with open('./Encoded features/encoded_trainfeature.pkl','wb') as f:
    pickle.dump(encoding_train,f)
```
